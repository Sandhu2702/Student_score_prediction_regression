```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#create database
data = {
    'Hours': [1, 2, 3, 4.5, 5, 6, 7, 8, 9, 10],
    'Scores': [17, 20, 27, 45, 50, 53, 65, 78, 85, 95]
}
df = pd.DataFrame(data)
```

```python
# 2. Visualize data
plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show()
```

```python
# 3. Split the data
X = df[['Hours']]
y = df['Scores']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 4. Train model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 5. Predict
y_pred = model.predict(X_test)
```

```python
# 6. Compare actual vs predicted
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(result)
```

```python
# 7. Plot regression line
line = model.coef_ * X + model.intercept_

plt.scatter(X, y, color='blue')
plt.plot(X, line, color='red')  # Regression line
plt.title('Regression Line (Hours vs Score)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show()
```

```python
# 8. Evaluate the model
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
# 9. Predict for a new value (e.g., 9.25 hrs)
hours = 8.25
predicted_score = model.predict([[hours]])
print(f"Predicted score if a student studies for {hours} hours: {predicted_score[0]:.2f}")
```
